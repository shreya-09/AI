Basics of using TensorFlow - How it works on prediction - Neural Network

```python
import tensorflow as tf
import numpy as np
```

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(units = 1)
])
```

```python
model.compile(optimizer='sgd' , loss='mean_squared_error')
```

```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype = float)
xy = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype = float)
```

```python
model.fit(xs, xy, epochs=500)
```

```python
print (f"model predicted : {model.predict(np.array([10.0]), verbose=0).item():5f}")
```
